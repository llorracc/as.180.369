## Abstract

This paper tries to investigate the rationality of consumer confidence through the lense of political party in power. First, we seek to establish a firm ground for the proposition that consumer sentiment even has the power to influence larger macroeconomic trends. Then, we undertake a literature review of the research done on the rationality of consumer sentiment. Next, we perform a regression analysis of economic sentiment from the University of Michigan Survey against several different indicators of political party in power to test a relationship between positive consumer sentiment and political party in control. We find that our model certainly establishes a correlation between Republican control of the senate and more positive consumer confidence. ## Introduction

Economic indicators play a vital role in nearly all forms of economic analysis and forecasting. Measurements like gross domestic product (GDP), unemployment, and inflation provide objective data to model trajectories of growth, business cycles, recessions and more. GDP, for example, represents the total value of goods and services produced in an economy over a set timeframe. Fluctuations in GDP directly inform the rate of a country's economic expansion or contraction. Gross Domestic Product is composed of four key components: personal consumption expenditures, private domestic investment, government spending, and net exports. According to {cite:t}`BEA-GDP-Shares`, personal consumption expenditures have accounted for over two-thirds of total GDP in the United States in recent years. Thus, consumer behavior and spending habits are hugely impactful in driving overall economic activity.

The primary indicator we use for both understanding and trying to forecast this consumer behavior is the index of consumer sentiment. In 1946, economist George Katona pioneered survey-based research examining the role of consumer psychology in shaping household spending decisions. Katona hypothesized that consumer behavior stems not just from objective facts about personal finances or the economy, but also the subjective perceptions, attitudes and expectations that individuals hold {cite:p}`um-survey-descript`. He developed sophisticated survey instruments to track consumer sentiment on topics ranging from personal finances, inflation and unemployment expectations, assessments of business conditions and attitudes toward major purchases. The University of Michigan Survey Research Center has carried out Katona's Surveys of Consumers monthly since 1946, gathering data from representative samples of American households on their consumer sentiment. Economists have in turn looked to the Survey's reports for indicators about the direction of the economy.

But the first question we should briefly answer is, are these economists right to do so? How well does consumer sentiment truly predict consumer spending?

From previous research done on the question, the answer seems to be yes. {cite:t}`consumer_carroll_1994` used a regression analysis model to show that the University of Michigan's Index of Consumer Sentiment (ICS) has considerable ability to forecast future changes in household spending, explaining about 14% of the variation in personal consumption growth. In fact, they concluded further that even though the precise reasons why remain unclear, that the ICS also has some incremental predictive power even controlling for other economic indicators like income growth, interest rates, unemployment, etc. Its forecasting ability is not just reflecting information contained in those other variables. These results align with findings by {cite:t}`consumer_ludvigson_2004`, who established the Consumer Sentiment Index as a statistically significant predictor of consumer demand. Through regression analysis, {cite:t}`consumer_ludvigson_2004` found consumer sentiment to be a key driver of consumption growth in the U.S., with a one-point rise being associated with a 1.3% increase in personal consumption. Other economists have found that the study has been a helpful tool in anticipating changes in other economic trends 6-9 months in advance, with a high correlation to outcomes like interest rate changes, inflation, and unemployment rate shifts {cite:p}`curtin_2019`. Thus, we can confidently say that consumer sentiment helps forecast household spending growth, and not just through its relationship with income or other indicators. 

With this general understanding in place, we can now move onto our second question.

Much of the predictive power of consumer sentiment hinges on one of the primary assumptions of economic theory in general: that consumers are rational.
But are consumers really rational when it comes to their sentiment about the economy?

In this paper,  we attempt to address this question. In the Literature Review section of this paper, we undertake a review of the literature surrounding the question of the role rational outlooks and decision-making plays for consumers in determining their economic sentiment. We discuss the methodology other research on the subject has utilized, and the results they derive. Then, in our Methodology section, we discuss our own attempt to build a model to shed some light on this question. We explain the data we used in our model and how the model takes shape. Finally, in our Results section we analyze the results our model provided and attempt to make some tentative conclusions about our research.


## Literature Review

The chief questions we investigate with this literature review are: how rational are consumers in determining their expectations for the economy? If they aren't, what influences their bias. 

First, a large body of the research done on the role of confidence in the economy takes either of two contrasting approaches. The first depends consumer confidence on the idea of "animal spirits"--coined by John Maynard Keynes {cite:p}`barsky_2012`. The 1990-1991 recession was directly preceded by a precipitous drop in all measures of consumer sentiment acccording to the ICS. Interestingly though, immediately before the sharp fall had been a period during which the sentiment index had reached record highs. {cite:t}`hall_1993` showed that after accounting for the actual changes to peoples' disposable incomes, the steep drop in sentiment and consumption during the recession was unwarranted. Instead, he theorizes that a series of negative sentiment instigated, perhaps, by the Iraqi invasion of Kuwait, led consumers to respond as if they had had a permanent decrease in their resources. They responded like this even though no objective metric besides for a brief oil-price spike would have indicated to them thus. As such, he observes that little of this fits in well to the neoclassical models that assume a 'rational agent'. Further, {cite:t}`blanchard_1993` not only points out that consumer confidence dropped steeply in response to the Iraqi invasion of Kuwait--which was a largely noneconomic event--but that it also remained significantly depressed for the next two years even after aggregate economic variables had rebounded. He attributes this, potentially, to an interpretation that steps outside of the usual utility-maximizing model, and posits that perhaps consumers fell prey to a sudden realization of overborrowing, or panic, etc. These reactions he calls "animal spirits" impulses. 

The second approach to consumer confidence, called by some the "news" or "information" view, suggests that movements in confidence reflect news about economic prospects received by consumers. In the Netherlands,{cite:t}`HOLLANDERS2011367` demonstrated that negative media coverage of the economy decreases consumer confidence, even when economic fundamentals remain unchanged. Through a time-series analysis, they found increased negative economic news coverage was associated with deteriorating consumer sentiment during the 1990s economic expansion. In the U.S. as well, {cite:t}`consumer_doms_2004` produced evidence supporting this view. Using an R-word index-- a list of words and phrases whose tone associates with economic downturn like "recession", "economic slowdown", etc.--they combed through prominent U.S. general news and economic news publications and found that U.S. consumers did indeed receive sentiment signalling from the tone and volume of the publications reporting on economic conditions. On an even deeper level, these signals pass through into consumer expectations indices and affect larger macroeconomic trends. Importantly, they showed that there have been times where the economic reporting ran contrary to what economic fundamentals actually reflect and thus consumer sentiment was pushed away from what they rationally should have been had they reflected true economic fundamentals. Similarly, {cite:t}`SHAPIRO2022221` utililzed lexicon-based techniques but also integrated machine learning techniques to analyze economic news texts and compare them to consumer sentiment releases. They found that the news sentiment was highly predictive of subsequent consumer sentiment data both positively and negatively. They also showed that positive sentiment shocks increased consumption, output, and interest rates while also temporarily reducing inflation. 

Another significant line of inquiry that researchers have made in this subject revolves around political affiliation affecting consumer sentiment. {cite:t}`political_deboef_2004` uncovered evidence that political partisanship significantly impacts consumer sentiment reports. In analyzing survey data, they found respondents identifying as Democrats reported substantially lower consumer sentiment following the 2000 election of President Bush, despite no notable changes in macroeconomic indicators. In a similar vain, {cite:t}`belief_meeuwis_2018` concluded that Republicans became more optimistic and increased the equity share of their portfolios relative to Democrats following the election. This effect holds even after controlling for differences in income, wealth, demographics, local economic conditions, etc. The portfolio reallocations are consistent with survey evidence showing Republicans became more optimistic and Democrats more pessimistic about future US economic performance after the election. Thus their findings suggest investors interpret public signals through the lens of their political identity, updating their beliefs and trading accordingly even though asset prices don't fully reflect their individual views. Although this research doesn't tie directly to consumer sentiment, it certainly corroborates the hypothesis that political considerations and identity even affects decisions of an economic nature.

In Australia, {cite:t}`effect_gillitzer_2018` show that Australian consumers report substantially higher economic sentiment when their preferred political party is in power compared to when the opposing party holds office in a way that can't be explained by underlying economic fundamentals or characteristics. This difference occurs immediately after elections which result in a change of power. The higher sentiment expressed by the affiliated group when their party gains power is then shown to translate into higher spending on cars relative to the opposing partisan group. This suggests the sentiment differences affect behavior. Thus, the sharp shifts in sentiment linked to election outcomes, the lack of corresponding economic changes at those precise times, and the subsequent relative spending changes, imply a degree of irrationality in the sense that consumers' views seem overly colored by partisan bias rather than objective conditions. Our research attempts to answer a similar but broader question: is there a certain political party that U.S. consumers, in general, feel more confident in. 


More generally, {cite:t}`katona_1975` concludes that consumer sentiment is broadly affected by the subjective state of mind of consumers. They posit that sentiment is a particularly useful metric during periods defined by extreme events since households’ expenditures depend not only on the ‘ability’, but also on the ‘willingness' to buy. Specifically, {cite:t}`waxman_2011` found that in Israel during the Second Intifada, even the perceived potential for a future terrorist attack lowered consumer sentiment and thus depressed consumer spending more broadly.  An interesting paper addresses the effect of terrorist attacks on U.S. consumer sentiment post 9-11. {cite:t}`terrorism_2015` finds empirical evidence that terrorist activity negatively impacts consumer confidence. They show via econometric analysis that their terrorism activity variable has a statistically significant negative coefficient in the model estimating consumer confidence. In addition, this finding only applies post-9/11. This supports the idea of an irrational overreaction specifically in the post 9/11 environment.  Thus, even though terrorism certainly has direct economic consequences on a given country, it is often the pass on effects generated by lower consumer sentiment into lower consumer spending through which terrorism most hurts an economy. 

Taken together, these studies indicate that while consumer sentiment gauges can strongly predict consumer behavior, the indices themselves may be impacted by individuals' subjective perceptions and biases. As a result, sentiment measures may overstate or understate the link between sentiment and future consumption. The findings highlight the need for caution in interpreting consumer confidence measures and underscore the complex psychological factors that influence people's economic outlooks and choices. More research is warranted to better understand these motivations driving consumer sentiment as well as how best to extract the true economic signal from noisy survey data. Integrating sentiment indices with other economic indicators and modeling techniques may allow for more accurate predictions of consumer behavior.




## Methodology


Our analysis looks at the potential link between consumer sentiment and Republican control of political institutions. The consumer sentiment data utilized in our analysis comes from the University of Michigan's Survey of Consumers. This survey has asked a national sample of about 500 households the same core questions every month since 1946, covering general sentiment, personal finances, business conditions, buying conditions, among other metrics of sentiment. By design, the samples used for the Survey are statistically representative of a cross-section of American households. Since 1960, this Survey has consistently utilized the same wording for questions with only minor changes in sampling and methodology. As a results, it offers us an unparalleled look at how U.S. consumer optimism has evolved since World War II.  

We specifically analyze the following data from the survey: 1. Current Sentiment Index; 2. Expected Sentiment 12 months from now; 3. Current Buying Conditions; 4. Current Personal Finance Sentiment; 5. Expected Personal Finance situation 12 months from now; 6. Business Conditions 12 Months from now. The reason for this is that these are the sole metrics with consistent quarterly data points for every year between 1960-2023. 

However, there are caveats associated with our choice of the Michigan data set. The survey slightly oversamples higher income households, which could skew optimism readings {cite:p}`consumer_bram_2007`. However, in aggregate over long periods, it represents one of the better and more accurate gauges of national consumer sentiment and confidence {cite:p}`um-survey-descript`.

To build our model, we gathered the aforementioned measures of consumer sentiment from 1960-2023 into a data file and averaged the quarterly data for each index into a yearly data point. In the same file, we attached an indicator of whether or not the Republican Party had control of the presidency, the senate, or the house of representatives in each year between 1960-2023. If the Republicans controlled any of the respective branches during a certain year, we marked a 1 in the corresponding cell, and if they didn't we marked a 0 in the cell. Then we ran OLS regressions with our columns indicating Republican control of the political bodies (the Presidency, the senate, the House of Reps.) as the independent variable, and the various sentiment indicators as our dependent, responding variable.

Thus, in total we ran 18 regressions. Further, in order to visualize and better grasp the trends indicated by our data, we created plots and filled in periods of GOP political power with red. 

There are a number of limitations to our model. Firstly, the majority of our familiarity with regression analysis came from a fairly cursory learning of the statistics behind it from interactions with the class TA and professor, as well as discussions with LLM's like ChatGPT and Claude.AI. In essence, the bulk of the code used in the analysis came from recommendations and fixes to our original code that were tendered by ChatGPT specifically. Secondly, our analysis looks primarily at correlation without deeper investigation into potential causal channels. Of course, many have heard the phrase correlation is not causation and thus it's difficult to infer causal relationships between political party in power and consumer sentiment from our study. There may also be issues with multicollinearity between political indicators that could weaken statistical interpretations. Another potential limitation is the phenomenon of multiple comparisons which pertains when multiple statistical tests are applied to a single set of data. Here we had multiple dependent values (e.g. current consumer sentiment, expected consumer sentiment, etc.) that were dependent on three distinct independent variables: each of the independent variables could be viewed as separate ‘trials’, whereas the number of dependent variables for each was 7 meaning the number of comparisons was 7 each of which then was tested separate for dependence on the three independent variables, (GOP_Pres, GOP_senate, etc.). So, the requisite probability was determined to be 0.0073 in lieu of 0.05 in order to yield an overall error of 5% for the set of comparisons. 

Nonetheless, this straightforward analysis offers a first pass look into whether any consistent relationships exist historically between Republican political dominance and consumer sentiment shifts across major household financial perspectives. Importantly, by using data that covers a long timeframe, we are better able to assess if relationships are temporally stable or show variation across decades. The fact that our data spans twelve presidencies, even more turnovers of the Senate and House of Reps., and multiple economic downturns, gives us a better     The results offer direction for where deeper dives into causal factors may be merited.



## Results

The main concept we sought to probe with this research was the potential for a relationship between various measures of U.S. consumer sentiment and the political party in power in the U.S. Presidency, Senate, and House of Representatives respectively. One serious caveat here is the writer's questionable familiarity with regression models and results, but our best effort was put forth in attempting to learn the relevant statistical indicators to focus on.

In general, our results were mixed. Briefly again, we ran a series regressions of Dependent-responding Y variables (current sentiment, expected sentiment, current personal finance sentiment, expected personal finance sentiment, expectations for business conditions over the next year and over the next 5 years, and current buying conditions) against three independent x variables (Republican President, Republican Senate, Republican House), respectively. 

Consistently, we found that from 1960-2023, the party of the president does not appear to be a major driver or predictor of consumer sentiment.
For example, when we regressed our Republican President flag against our consumer sentiment indices, our R-squared values were consisently very low (0.001, 0.000, 0.004, etc.) indicating that only tenths of a percent of the variation in our consumer sentiment indices were explained by the the sitting of a Republican President. In fact, many of our adjusted R-squared's were actually negative, further reinforcing that the model has very little explanatory power.
Further, the p-values for these regressions were much higher than the typical 0.05 or 0.01 significance level, consistenly hovering above 0.5 and sometimes even approaching 0.9. This means we cannot reject the null hypothesis that there is no relationship between a GOP president and consumer sentiment. Our coefficients were small and often our 95% confidence intervals crossed zero further indicating a lack of statistical significance. 
Thus, there is no evidence of a statistically significant relationship between Republican control of the Presidency and the various measures of consumer sentiment. 

There are a few potential reasons that this could be. One is that consumer sentiment is influenced more by economic factors than by politics. Certainly, because the ICS attempts to interview a representative sample of American households, it woul be reasonable to presume that the political affiliations of the sample members do not skew too strongly in either political direction. Therefore, we would assume that the consumers' sentiment won't especially favor one party's president over another's. Thus, things like employment rates, inflation, stock market performance, etc. tend to drive consumer confidence to a greater extent than political factors. A second reason is that most modern U.S. presidents, regardless of party affiliation, have not pursued drastically different economic policies. For example, the economy did very well under the Presidency of Bill Clinton--a social democrat--even though he didn’t pursue policies so different to those of other Democratic presidents between 1960-2023. Even so, consumer sentiment was much higher than it was for other Democratic presidents. Thus, there may be little discernible difference in outcomes. Finally, the business cycles affected by a president's policies often do not coincide with their presidency. According to {cite:t}`https://doi.org/10.1111/ajps.12145`, sometimes consumers reward politicians with positive sentiment for where the economy will likely go because of their policies and sometimes they dont. Thus, it is difficult to parse the true effects any president will have on sentiment about the economy.


On the other hand, however, most of our regressions yielded much more promising results when we tested the same relationships but with a GOP majority senate instead of a GOP president. By and large, our R-squared values ranged from the mid-teens to mid-twenties percentiles with the exceptions of our regressions against future business conditions and as a result the expected consumer index. Thus, we found that Republican control of the Senate definitely accounts for some of the variation in the consumer sentiment index over our range of dates. 
Additionally, the coefficients for these regressions were consistently positive and most of them represented fairly significant boosts in consumer sentiment associated with Republican senates. These coefficients were generally statistically significant, since they often corresponded with exceedingly low p-value's. As a result, looking at the results from our regressions alone, one would think that there is a significant positive relationship between Republican Senate control and more positive personal finance sentiment. And there is. But when we analyze a few of the graphs of our most promising regressions, a different picture begins to take shape. 






````python
##plot of CurPersonalFinance vs. Senate

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('Sentiment1960_plot.csv')


# Create a line plot with economic indicator values
plt.plot(df['Year'], df['Personal Finance Current'], label='Personal Finance Current', linestyle='-', color='b')

# Shade the periods with Republican majority (where the column is 1)
plt.fill_between(df['Year'], 0, df['Personal Finance Current'], where=(df['GOP_senate'] == 1), color='red', alpha=0.3, label='Republican Senate Majority')

# Set labels for the axes
plt.xlabel('Year')
plt.ylabel('Personal Finance Current')

# Set the title of the plot
plt.title('Notice the fact that Democrat-majority senates coincided with beyond their control financial crises like in 1990-1991, and 2008')

# Add a grid
plt.grid(True)

# Show the legend
plt.legend()

# Show the plot
plt.show()
````

We can glean a few very important things from this graph. First, we can see that during the recession of 1990-1991 the Senate was under Democratic control. According to {cite:t}`blanchard_1993`, it's likely that one of the primary drivers of this recession was the Iraqi invasion of Kuwait, an event that the U.S. Senate certainly had little control over. As a result the Democratic senate was caught holding the proverbial bag, and poor consumer sentiment during this period is likely not a reflection of American consumers making a judgement about the Democratic control of the Senate. Secondly, in 2008, Democrats were again in control of the Senate during a financial emergency that they had little to no responsibility for. In fact, it is more likely that voters had issued a negative referendum on the Republican majority responsible for the Iraq war and thus had voted in a Democratic senate that would quickly be greeted by a very unpleasant financial situation. At root, though, this graph helps illustrate the likelihood that the positive correlation between Republican Senates and consumer confidence is likely a coincidence.

As was the case with our regressions comparing Republican presidents to consumer confidence, we did not find a statistically significant relationship between Republican control of the House of Representatives and American consumer confidence.
