# Comments on Classmates' Reaction on Korinek Paper

I found Bruce’s approach to using ChatGPT particularly inspiring, as he showed how AI can take care of small, repetitive tasks that are too minor to justify human effort, allowing researchers to dedicate more time to creative thinking and interpretation. This idea connects well with Elvis’s approach, where ChatGPT was mainly used for managing citations—a task that usually takes up a lot of time and attention but doesn’t necessarily contribute to the deeper analytical parts of research. What stood out to me most, however, was how Bruce took this one step further by using ChatGPT to test and refine the setup of hypotheses before actually running them in Python. That struck me as a clever way of integrating AI into the early, conceptual stage of quantitative research, where the model can help check logic, spot inconsistencies, or even simulate expected relationships. I also liked the idea of breaking the research process into smaller, manageable sections with ChatGPT’s help. This makes the entire workflow more flexible and less overwhelming—especially when revising essays or refining arguments. Instead of rewriting large portions at once, one can focus on improving specific parts with immediate feedback, which mirrors how real academic writing and experimentation evolve in stages. Overall, both Bruce’s and Elvis’s examples show how AI can become a collaborative tool—not replacing researchers, but helping them think more clearly, organize more efficiently, and experiment more confidently.
