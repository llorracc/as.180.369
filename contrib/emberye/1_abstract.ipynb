# Abstract

This paper investigates the relationship between vocational training (VCT) investments and macroeconomic outcomes, with a focus on both GDP per capita and economic resilience during the COVID-19 pandemic. Initially, the study assesses whether higher investment intensity in vocational training, reflecting the cost and quality of human capital development, is positively associated with a countryâ€™s GDP per capita, using the Fresher coefficient to measure correlation. Building on these findings, the analysis then explores whether greater VCT investment contributes to enhanced economic resilience, operationalized as lower GDP losses during the COVID-19 shock. Depending on the statistical significance of the results, the study may report either correlations or, where sufficient, regression estimates quantifying the extent to which increases in VCT investment correspond to improvements in economic resilience, measured in percentage points. The findings aim to provide empirical evidence on the dual role of vocational training in fostering both economic growth and crisis mitigation.
