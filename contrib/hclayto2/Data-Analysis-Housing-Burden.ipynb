The Rose Foundation compiled a list of all CEQA lawsuits filed between 2019-2021, with a table giving the details of each lawsuit (pp. 65-99, "CEQA by the Numbers"). Lawsuits are broken down by type of development affected and number of housing units, if applicable. I was able to identify some major potential case studies, such as a May 2019 lawsuit filed in Los Angeles by the organization Climate Resolve targeting a proposed development of nearly 20,000 mix-used units. Two lawsuits similarly targeting mixed use housing developments in San Diego and Sacramento were also filed in 2020, with 3000 or more units in each of these proposed developments. Both of these latter two projects were targeted by EIRs. A 2021 lawsuit targeted a more than 1,000 unit mixed-use housing development project in Los Angeles. It is important to note that most lawsuits filed in this period were not directed at housing development projects. 

The 2019 lawsuit seemed particularly significant, and after searching for it online I was able to identify a case file created by Columbia Law School. The file indicates that a settlement was reached in late 2021 enabling the project to go ahead, but only after a 2.5 year process. This may be a good case study to zero-in on, because the project was so big, potentially being large enough to affect housing prices in the area. However, since the lawsuit was only resolved in 2021, the effects may be hard to gauge.
