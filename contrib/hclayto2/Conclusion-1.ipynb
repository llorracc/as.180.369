# Conclusion

In this paper, I have examined the effects of the California Environmental Quality Act (CEQA) on housing prices in California. In doing so, I have attempted to inform a growing debate about the effects of this law on the state's affordable housing crisis, especially in the wake of recent reforms to the law. By conducting a comprehensive review of the development of housing prices over time in different areas of the state, particularly in areas where CEQA has been most intensively applied to housing projects, I have attempted to identify the effects of this law on housing prices. My main sources of data have been the Federal Housing Finance Agency's Housing Price Index (HPI) and a comprehensive record of CEQA lawsuits compiled from various sources. 

Overall, my review indicates that, where CEQA lawsuits have been most common and where they have affected large housing projects, housing prices have risen faster than statewide averages. However, establishing a definitive causal relationship between CEQA and rising housing prices will take more research. Furthermore, as numerous sources make clear, the rate of CEQA intervention against housing developments in the state is relatively low. The law is most commonly applied against other types of infrastructure and building development, and in the case of housing, most projects are not affected by a CEQA lawsuit. Nevertheless, there is some evidence suggesting that CEQA has played a role in raising housing prices in jurisdictions where it has been most commonly invoked. 

This review has focused on CEQA lawsuits targeted at housing projects in order to discern the effect on housing prices. Further research on CEQA's housing market effects would ideally include CEQA lawsuits targeted at non-housing developments as well, since the law's effects on these projects could also have an effect on housing prices. 

My study provides a new empirical approach for assessing CEQA's effects and builds on methods used by researchers of environmental policy and housing. The existing literature on CEQA has primarily made use of survey data and has been conducted by environmental interest groups with an stake in the law's preservation. While this literature provides a useful starting point, it is important that independent economics researchers undertake studies of the law in order to deduce its full effects.
