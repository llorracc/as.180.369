# Introduction

The California Environmental Quality Act (CEQA) is a landmark, 1970 environmental protection law signed. Passed in the early days of the modern environmentalist movement, the law has inspired similar environmental legislation in states around the country. Specifically, the law intended to insure accurate assesment and disclosure of the environmental impacts of proposed public development projects, thus protecting Californians and the environment from the deleterious effects of construction. The law was soon expanded through a series of court decisions to apply to most private projects as well. The central component of the law is environmental review, in which a detailed assesment of any proposed new development is required for the project to go forward. 

The law has made California a leader in environmental protection, but has also in recent years come to be blamed for one of the state's most pressing challenges: its housing affordability crisis. California has the highest homelessness rate in the nation and an overall crisis of expensive and limited housing. CEQA, with its detailed, lengthy review process, and avenues for assessments to be challenged by concerned parties in court, has led many to identify it as one of the chief culprits of California's housing crisis. Responding to this crisis, California lawmakers significantly scaled back CEQA in 2025, in bipartisan fashion. An environmental law signed into law by a Republican governor Ronald Reagan had come to be blamed by leaders across the political spectrum for the economic failures of the state. Given the recent repeals, it is important that we assess closely CEQA's effects on housing prices, especially as California leaders weigh scaling it back further and as states around the country consider similar reform of environmental law to adress their housing crises. In this paper, I will analyze the effects of CEQA litigation on housing prices in the state of California.
