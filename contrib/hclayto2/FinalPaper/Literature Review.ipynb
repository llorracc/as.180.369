# Literature Review

According to my review of the literature, there is a  significant economic research on the effects of environmental regulations on the housing market, but little economic research on CEQA's effects specifically. The economics literature explores both the demand- and supply-side effects of environmental regulations on the housing market, though there is particular attention to the demand-side of the market. Specifically, a large number of studies demonstrate how environmental regulations, by improving the quality of surrounding areas, drive up housing prices. The "hedonic pricing" model isolates the impacts of specific, government-protected environmental amenities, such as clean air and preserved natural surroundings, on housing prices. 

A 2004 review of the environmental economics and housing literature by Katherine Kiel covers studies that consider the effects of national, state, and local environmental laws from the Clean Air Act to "smart growth" and land-use policies. This review provides a good introduction to studies that demonstrate the  demand- and supply-side effects of environmental regulations on housing prices, though it also demonstrates the more limited supply-side research. Using this literature review, I was able to identify a number of key studies as well as methodologies used by economists in the field. One study by Benton and Pollack (1992) looked at the economic effects of land-use environmental protection policies. The authors focused their study on the effects of Maryland's Critical Area Law and found that the law had a positive overall effect on the market, as vacant land values held steady and residential housing values rose. This study shows how environmental policy can be crafted to balance environmental protection with developmental needs. Chay and Greenstone (2005) use a difference-in-differerences approach with instrumental variables to show how reductions in air pollution from the Clean Air Act raised housing values at the county level. Keiser and Shapiro (2019) demonstrate the mixed property market effects of the Clean Water Act; Hanniger, Ma and Timmins (2017) show how the remediation of contaminated lands raises property values; and Atreya, Ferreira, and Kriesel (2013) show how improved flood risk maps increase housing values. Overall, this literature tends to focus on how environmental regulations increase demand for housing and thereby prices. This literature provides a good point for contextualizing my study of CEQA and serves as inspiration for the difference-in-differences method I use in my study. However, it also shows the need for more supply-side effects of environmental regulations on housing prices.

Focusing on CEQA-housing literature specifically, most of the research in this area has been done by scholars not in the field of economics. For example, legal scholar Jennifer Hernandez conducted an ambitious, three-part study of CEQA litigation, showing how it slowed housing development and increased project costs across the state of California. Other legal studies, such as O'Neill, Biber and Marantz (2023) and Elmendorf, Biber, Monkonnen and O'Neill (2020) have reviewed the impacts of land-use policies such as CEQA on California's affordable housing crisis and advanced suggestions for reform. The Berkely Terner Center for Housing Innovation, Public Policy Institute of California (PPIC), UCLA Institute of Transportation Studies, and other California-based academic institutes and think tanks also frequently release memos and studies on CEQA. Similarly, a range of environmental advocacy organizations, such as the Association of Environmental Professionals (AEP), have commissioned studies on CEQA, with these groups often calling into question its negative effects. However, all of these studies are rarely grounded in economics research methods.

Overall, there is a substantial economics literature on environmental policy and housing, and a significant literature on CEQA, though very few studies that bridge these two domains. That is, there is very little economics literature on the housing market effects of CEQA specifically. In this paper, I will attempt to begin bridging this gap.  


## Additional:


For my research, I will be focusing on the effects of the California Environmental Quality Act, or CEQA. In 2018, the California Association of Environmental Professionals (AEP), "a statewide resource representing environmental professionals," commissioned a study on this law to measure its effects on the housing crisis in the state and explore avenues for possible reform. In the introduction, the authors note that, "To date, no comprehensive data about CEQA review methods including use of exemptions and streamlining options has been published." This indicates that there is in fact a lack of empirical research on the issue. The study involved a survey of all cities and counties in the state and, though only garnering a 9% response rate, disproportionately received responses from high-population jurisdictions with significant housing development. Overall, the study found that "Streamlining/Exemptions was the predominant type of environmental review used
for housing projects in respondent jurisdictions (42% of projects), followed by
Mitigated Negative Declarations (36% of projects). Only 6% of projects were reviewed
by EIRs [Environmental Impact Reports]." This finding indicates that burdensome CEQA reviews, in the form of EIRs, were not that prevalent. Therefore, the authors' findings shed doubt on the claim that CEQA EIRs are a major force in blocking needed housing construction. However, they also found that affordable housing projects faced a hard time securing exemptions and were more likely to be subjected to EIRs, indicating that CEQA comes down disproportionately hard on affordable housing projects. Furthermore, the authors conclude, "The perceived causal relationship between CEQA review and frequent project withdrawals may be significantly overstated." Overal, this study, while helpful, should be taken with a grain of salt due to its commission by an environmental group with an interest in protecting CEQA.
