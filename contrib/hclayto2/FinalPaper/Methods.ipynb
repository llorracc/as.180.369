# Methods

## Research Design

This study seeks to evaluate the effects of CEQA litigation on housing prices in the state of California. Specifically, it focuses lawsuits filed to challenge environmental reviews of proposed housing projects.

## Data Sources

I use two primary sources of data for this study. The first is a set of all lawsuits filed under CEQA in the  three-year period from 2019 to 2021. This data was compiled by the __Rose Foundation__, a California-based environmental non-profit group, as part of a 2023 study. It is the most comprehensive, publicly available data set of CEQA litigation. These lawsuits are filed to challenge government-mandated environmental reviews of proposed construction or development projects. I filtered this data to only include lawsuits targeting housing projects. Overall, there were 109 CEQA lawsuits targeting housing projects in California in the three-year period under study. While non-housing CEQA lawsuits may indirectly affect housing prices in California, an analysis of such lawsuits is outside the scope of this study.

The second data set I use is the __Federal Housing Finance Administration's Housing Price Index (HPI)__. HPI data is made publicly available and charts the development of housing prices all around the country at a variety of jurisdictional levels. For the purposes of this study, I filtered HPI data at the county level in California. 

Other data sources include 

* __Columbia Law School Climate Litigation Database__: keeps track of major lawsuits related to climate issues, including CEQA lawsuits, and provides a timeline and detailed analysis of each lawsuit; helpful in charting the development of individual lawsuits for the purposes of the DiD analysis

## Empirical Approach

### Regression Analysis

I begin by conducting a simple linear regression analysis, comparing the percentage rise in HPI over different time periods to the intensity of CEQA litigation. For each county, I constructed a __CEQA litigation intensity index__ measuring the number of housing units affected per 100k people in each county. This index was derived by first calculating the total number of housing units affected by CEQA litigation from 2019 to 2021 in each county. For each county, I divided this number by the county population according to 2018 census estimates and then multiplied by 100k to get the number of units affected per 100k people.

CEQA litigation intensity = (total # of housing units affected by litigation from 2019-2021) / (2018 population) * 100k

After doing this for each county, I calculated the percentage rise in HPI in each county, first from 2015 to 2024, and then from 2019 to 2024.

### Difference-in-Differences 

With the recently approved CEQA reforms of 2025, it is worth considering the effects of past CEQA reforms on housing prices in California. California has passed a patchwork of reforms to CEQA in recent years in an effort to adress the state's climate and housing goals. One of the most significant reforms was Senate Bill (SB) 743, which came into effect statewide at the beginning of Q3 2020. This reform fundamentally changed how the state measures the transportation impacts of projects under CEQA review by replacing traditional traffic congestion metrics (like Level of Service or LOS) with Vehicle Miles Traveled (VMT). In the past, projects were assessed according to how much traffic congestion they created, which incentivized greater urban sprawl, more and wider roads, and less dense housing developments. With the change to VMT in 2018, projects were now incentivized to minimize how much their residents would have to use cars to travel. This in turn incentivized more compact development patterns and more urban infill housing. As a result, I hypothesize that the transition from LOS to VMT under SB 743 would lead to more and denser urban housing and thus slow the rise in housing prices. 

To test this hypothesis, I conduct a difference-in-differences analysis. I begin by assembling all quarterly HPI data on twenty-three metropolitan statistical areas (MSAs) in California. I then code each MSA according to its pre-2020 LOS exposure. Some MSAs were highly affected by the LOS transportation impact requirements, whereas other MSAs were not. MSAs with higher pre-2020 LOS exposure are hypothesized to be more affected by the transition to VMT in 2020 and thus be more likely to exhibit a change in housing prices.
