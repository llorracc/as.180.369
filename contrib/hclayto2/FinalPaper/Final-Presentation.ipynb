# Introduction

    * Faster legal review
    * Easier rezoning

# Results - LA County

* Prices rose faster in areas of higher CEQA litigation
* LA County: HPI rose ~38% from 2019 to 2022; it rose ~27% in California
    * Climate Resolve v. County of Los Angeles (20k units)


# Results - San Diego County

* As in LA county, there was a high level of CEQA litigation
* HPI rose ~39% from 2019 to 2022
* Preserve Wild Santee v. City of Santee (3k units)

# Results - Sacramento County

* 3 lawsuits of 3,000 or more units
* HPI rose ~39% from 2019-2022

# Conclusion

* Prices rise more (and more quickly) in jurisdictions with higher rates of CEQA litigation

* Causal relationship?
    * Issue of the pandemic
    * Litigation concentrated in counties that are already the most highly populated and in-demand
    * Sheer # of total housing units vs. relatively small # affected by CEQA litigation
    * However, as Hernandez (2022) points out: only 110k units of production a year
