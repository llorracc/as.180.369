# Abstract
This paper examines the relationship between environmental regulation and housing prices by studying the California Environmental Quality Act (CEQA). Drawing on insights from the urban and environmental economics literature on regulatory constraints and housing supply, the analysis proceeds in two parts. First, I construct a county-level measure of CEQA litigation intensity based on housing projects affected by litigation between 2019 and 2021 and examine its association with housing price growth using data from the Federal Housing Finance Agency. Second, I evaluate the effects of a major CEQA reform—Senate Bill 743, which replaced congestion-based transportation review with a Vehicle Miles Traveled standard—using a difference-in-differences design across California metropolitan statistical areas. The results reveal little evidence of a strong relationship between CEQA litigation intensity and housing price growth at the county level, and only modest, statistically imprecise effects of the SB 743 reform on housing prices. Together, these findings suggest that CEQA-related constraints do not generate large, readily detectable impacts on housing prices over the study period, while highlighting the need for more granular, causally identified evidence on the supply-side effects of environmental review regimes.
