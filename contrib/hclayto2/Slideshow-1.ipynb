# The California Environmental Quality Act (CEQA) and Housing Prices

# Introduction 

* Environmental policy and housing markets: a real, though relatively small, field of economics 
* What is the __California Environmental Quality Act (CEQA)__?
    * A landmark environmental protection law passed in 1970 (signed by __Ronald Reagan__!)
      * Many other states followed suit with similar laws (e.g., New York and Massachusetts)
    * Key provisions
      * Environmental review
      * __Negative Declarations__ (NDs) or __Environmental Impact Reports__ (EIRs)
      * Exemptions and streamlining (emergency projects, home expansions, certain affordable housing projects)
* California has become a leader in environmental protection -- but at what cost?
    * __Blamed for rising housing prices__
* Significantly scaled back in summer 2025, with bipartisan support (!)
    * Exempt high-density projects from stringent review
    * Accelerate legal review
    * Exempt numerous types of development
    * Easier rezoning
* All these changes will make it harder for all types of groups, organizations and individuals to file potentially costly and time-consuming lawsuits -- <mark>my focus</mark>
* Highly relevant and in urgent need of research
* How do we approach this topic?

# Results 

* Prices rising faster in areas of higher CEQA litigation
    * For example, in Los Angeles County the HPI rose about 38% between 2019 and 2022 whereas it rose about 27% in California as a whole
    * This was an area of especially high CEQA litigation


# Results (cont.)

* Causal relationship?
    * Issue of the pandemic
    * Hard to say because of sheer number of total housing units vs. the relatively small number affected by CEQA litigation
    * Also, CEQA litigation concentrated in counties that are already the most highly populated and in-demand

# Conclusion

* Prices rise more (and more quickly) in jurisdictions with higher rates of CEQA litigation
* More research needed into effects of non-housing CEQA litigation
* Is CEQA a big culprit in California's housing crisis?
    * Results are mixed
