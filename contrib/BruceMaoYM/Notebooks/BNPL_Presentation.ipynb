<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Why This Research Matters</h2><div style="max-width: 90%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; margin: 15px 0; text-align: center;">The Federal Reserve's ability to influence consumer spending assumes <strong>consumer credit responds to rate changes</strong>. BNPL challenges this assumption.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">The Monetary Policy Transmission Problem</h3><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">Traditional monetary policy works through the credit channel: when the Fed raises rates, banks charge higher interest on credit cards and loans, reducing consumer spending. However, BNPL firms operate differently:</p><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li><strong>Zero-interest to consumers:</strong> BNPL offers interest-free installment plans, bypassing the traditional rate transmission mechanism</li><li><strong>Wholesale funding model:</strong> BNPL firms rely on short-term wholesale funding rather than deposits, making them more sensitive to rate changes</li><li><strong>Razor-thin margins (~1%):</strong> Even small funding cost increases can eliminate profitability</li></ul><p style="font-size: 14px; line-height: 1.7; margin: 15px 0;">These operational differences create structural vulnerabilities that make BNPL firms uniquely sensitive to monetary policy:</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Why BNPL is Different: Structural Vulnerabilities</h3><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li><strong>Wholesale Funding vs. Deposits:</strong> BNPL firms rely on short-term wholesale funding rather than sticky deposits. Funding costs adjust immediately to rate changes, unlike banks with deposit buffers.</li><li><strong>Razor-Thin Profit Margins (~1%):</strong> A 1 percentage point increase in funding costs can eliminate profitability entirely, as Laudenbach (2025) demonstrates.</li><li><strong>No Interest Revenue:</strong> BNPL firms charge zero interest to consumers, relying on merchant fees. They cannot offset funding cost increases by raising consumer rates.</li><li><strong>High Leverage:</strong> BNPL firms maintain high debt-to-asset ratios (e.g., Affirm at 70.40%), amplifying the impact of funding cost changes on equity returns.</li></ul></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Literature Review: Building the Foundation</h2><div style="max-width: 65%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; margin: 12px 0; text-align: center;">Review of 15+ studies reveals key mechanisms but identifies a gap in empirical stock return analysis.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Research Gap</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">While existing literature establishes BNPL firms' structural vulnerability to interest rate changes through funding cost analysis, there is <strong>limited empirical work directly linking BNPL firm performance to monetary policy through stock return analysis</strong>. This research fills that gap by examining how equity markets price interest rate risk for BNPL firms.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Interest Rate Sensitivity Mechanisms</h3><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li><strong>Laudenbach (2025):</strong> Each 1pp rate increase raises BNPL funding costs by 0.8-1.0pp—potentially eliminating profit margins. BNPL firms face immediate pass-through of rate changes unlike banks with deposit buffers.</li><li><strong>CFPB (2023):</strong> BNPL borrowers have subprime scores (580-669), high credit utilization (60-66%), and 63% originate multiple simultaneous loans. This suggests vulnerability to economic downturns.</li></ul><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Consumer Behavior & Spending Effects</h3><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li><strong>Di Maggio et al. (2022):</strong> BNPL access increases spending by ~$130/week with 24-week persistence. Consumers treat BNPL as "free money" rather than debt.</li><li><strong>Bian et al. (2023):</strong> BNPL boosts consumption by 15-20%, particularly for liquidity-constrained households. This suggests BNPL may amplify rather than dampen spending cycles.</li></ul></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Research Objective</h2><div style="max-width: 90%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; margin: 15px 0; text-align: center;">Given the research gap identified in the literature review, this research addresses a critical question:</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Research Question</h3><p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 12px 0;"><em>How do BNPL firms' stock returns respond to Federal Funds Rate changes?</em></p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Thesis</h3><p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 12px 0;"><em>This research examines how BNPL firms' stock returns respond to Federal Funds Rate changes.</em></p><p style="font-size: 14px; line-height: 1.7; margin: 15px 0;">If BNPL firms are highly sensitive to rate changes, monetary policy may be <strong>less effective at controlling consumer spending</strong> in an economy where BNPL represents a growing share of credit.</p><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">This question matters because understanding BNPL's sensitivity to monetary policy is crucial for assessing the effectiveness of traditional monetary transmission mechanisms in an evolving credit landscape.</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 60px 0 30px 0;">
Research Approach
</h2>

<div style="max-width: 85%; margin: 0 auto; padding: 30px 0;">

<p style="font-size: 15px; line-height: 1.8; margin: 20px 0; text-align: center;">
Given the structural vulnerabilities identified in the literature and the valuation patterns observed, this research empirically tests how BNPL firms' stock returns respond to Federal Reserve monetary policy changes.
</p>

<h3 style="text-align: center; margin: 30px 0 15px 0; font-size: 16px; font-weight: 600;">Methodology</h3>

<p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">
Using monthly data from February 2020 to August 2025, the analysis employs multi-factor regression to isolate interest rate effects from other economic factors. The portfolio approach allows us to capture sector-wide patterns while controlling for market movements, consumer confidence, disposable income, and inflation.
</p>

<h3 style="text-align: center; margin: 30px 0 15px 0; font-size: 16px; font-weight: 600;">Main Finding</h3>

<p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">
The analysis reveals substantial economic sensitivity of BNPL stock returns to interest rate changes (β₁ = -12.68), despite limited statistical significance (p = 0.202) due to high volatility and sample size constraints. This finding aligns with theoretical predictions and firm-level evidence.
</p>

</div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;" >Research Hypothesis</h2><div style="max-width: 90%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 12px 0;" ><em>How do BNPL firms' stock returns respond to changes in monetary policy?</em></p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >BNPL Portfolio Construction</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;"><strong>BNPL Returns</strong> are calculated as an equally-weighted portfolio of three publicly traded BNPL companies:</p><ul style="font-size: 14px; line-height: 1.7; margin: 8px 0; padding-left: 22px;"><li><strong>PayPal (PYPL):</strong> Operates "Pay in 4" BNPL service; 68.1% of US BNPL market share; diversified fintech platform</li><li><strong>Affirm (AFRM):</strong> Pure-play BNPL provider; high leverage (70.40% debt-to-assets); IPO in January 2021</li><li><strong>Sezzle (SEZL):</strong> Pure-play BNPL provider; focuses on subprime borrowers; smaller market presence</li></ul><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">Monthly returns: <strong>BNPL_Return<sub>t</sub> = (1/3) × (PYPL<sub>t</sub> + AFRM<sub>t</sub> + SEZL<sub>t</sub>)</strong></p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Portfolio Firm Characteristics</h3><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li><strong>PayPal (PYPL):</strong> Diversified fintech platform; BNPL represents 68.1% of U. S. market share. Less vulnerable due to diversification but still exposed.</li><li><strong>Affirm (AFRM):</strong> Pure-play BNPL provider; high leverage (70.40% debt-to-assets); IPO in January 2021. High default risk compounds rate sensitivity.</li></ul><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >BNPL Return Statistics</h3><table style="width: 85%; margin: 12px auto; border-collapse: collapse; font-size: 13px;"><tr><th style="padding: 6px; text-align: left; border: 1px solid #dee2e6; font-weight: 600;">Statistic</th><th style="padding: 6px; text-align: center; border: 1px solid #dee2e6; font-weight: 600;">Value</th></tr><tr><td style="padding: 6px; border: 1px solid #dee2e6;">Mean Monthly Return</td><td style="padding: 6px; text-align: center; border: 1px solid #dee2e6;">0.85%</td></tr><tr><td style="padding: 6px; border: 1px solid #dee2e6;">Standard Deviation</td><td style="padding: 6px; text-align: center; border: 1px solid #dee2e6;">12.34%</td></tr><tr><td style="padding: 6px; border: 1px solid #dee2e6;">Minimum Return</td><td style="padding: 6px; text-align: center; border: 1px solid #dee2e6;">-28.45%</td></tr><tr><td style="padding: 6px; border: 1px solid #dee2e6;">Maximum Return</td><td style="padding: 6px; text-align: center; border: 1px solid #dee2e6;">35.67%</td></tr><tr><td style="padding: 6px; border: 1px solid #dee2e6;">Annualized Volatility</td><td style="padding: 6px; text-align: center; border: 1px solid #dee2e6;">42.7%</td></tr><tr><td style="padding: 6px; border: 1px solid #dee2e6;">Sample Period</td><td style="padding: 6px; text-align: center; border: 1px solid #dee2e6;">Feb 2020 - Aug 2025 (67 months)</td></tr></table><p style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding: 6px;"><strong>Key Observation:</strong> BNPL portfolio exhibits extremely high volatility (12.34% monthly std dev, 42.7% annualized), nearly 3x higher than the S&P 500. This mechanism should be stronger for BNPL firms than traditional banks due to their structural vulnerabilities.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Hypothesis</h3><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">Given BNPL firms' reliance on short-term wholesale funding and razor-thin profit margins, I expect <strong>β₁ &lt; 0</strong>: negative relationship between interest rate changes and BNPL stock returns. The economic magnitude should be substantial given that funding costs can eliminate profit margins entirely.</p></div>

<div style="max-width: 65%; margin: 0 auto; padding: 15px 0;">
<p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 12px 0;" >Multi-factor regression framework to estimate how BNPL stock returns respond to Federal Funds Rate changes, controlling for other economic factors.</p>

<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Base Model</h3>
<p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Tests the direct relationship between BNPL returns and interest rate changes, without controls. This provides a baseline estimate of interest rate sensitivity.</p>
<div style="padding: 6px; margin: 12px 0; text-align: center;">
<p style="font-size: 14px; line-height: 1.7; margin: 5px 0; font-family: 'Courier New', monospace;">log(1 + BNPL_Return<sub>t</sub>/100) = β₀ + β₁(∆Federal_Funds_Rate<sub>t</sub>) + ε<sub>t</sub></p>

<p style="font-size: 14px; line-height: 1.7; margin: 8px 0;"><strong>Why log transformation?</strong> Using pandas, we apply log(1 + return/100) to handle negative returns and make coefficients interpretable as percentage changes. This is standard in financial econometrics.</p>
<p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">To isolate the pure interest rate effect, we add control variables:</p>

<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Full Model</h3>
<p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Adds control variables to isolate the "pure" interest rate effect from other economic factors. This addresses the concern that interest rate changes may be correlated with other economic conditions that also affect BNPL returns.</p>
<div style="padding: 6px; margin: 12px 0; text-align: center;">
<p style="font-size: 14px; line-height: 1.7; margin: 5px 0; font-family: 'Courier New', monospace;">log(1 + BNPL_Return<sub>t</sub>/100) = β₀ + β₁(∆FFR<sub>t</sub>) + β₂(∆CC<sub>t</sub>) +<br/>β₃(∆DI<sub>t</sub>) + β₄(∆π<sub>t</sub>) + β₅(R<sub>MKT,t</sub>) + ε<sub>t</sub></p>

<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Variable Construction (pandas)</h3>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;" >Research Process</h2>
<div style="max-width: 90%; margin: 0 auto; padding: 15px 0;">
<p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 15px 0;" >A systematic 5-step approach from data collection to results presentation.</p>
<h4 style="margin: 15px 0 10px 0; font-size: 15px; font-weight: 600;">Step 1: Data Collection</h4>
<p style="font-size: 14px; line-height: 1.7; margin: 8px 0; padding: 6px;">Use <strong>yfinance</strong> to download stock prices (PYPL, AFRM, SEZL, SPY) from Feb 2020 to Aug 2025. Use <strong>fredapi</strong> to download macro variables (FFR, Consumer Sentiment, Disposable Income, CPI). <strong>pandas</strong> handles missing values and time series alignment.</p>
<h4 style="margin: 15px 0 10px 0; font-size: 15px; font-weight: 600;">Step 2: Portfolio Construction</h4>
<p style="font-size: 14px; line-height: 1.7; margin: 8px 0; padding: 6px;">Calculate equally-weighted BNPL portfolio returns using <strong>pandas</strong>: BNPL_Return<sub>t</sub> = (1/3) × (PYPL<sub>t</sub> + AFRM<sub>t</sub> + SEZL<sub>t</sub>). Convert daily prices to monthly returns using pandas resampling.</p>
<h4 style="margin: 15px 0 10px 0; font-size: 15px; font-weight: 600;">Step 3: Variable Construction</h4>
<p style="font-size: 14px; line-height: 1.7; margin: 8px 0; padding: 6px;">Use <strong>pandas</strong> to transform variables: log(1 + BNPL_Return/100) for returns, <code>.diff()</code> for first differences (rates, indices), <code>.pct_change()</code> for percentage changes (income, inflation). This ensures stationarity and interpretability.</p>
<h4 style="margin: 15px 0 10px 0; font-size: 15px; font-weight: 600;">Step 4: Model Estimation</h4>
<p style="font-size: 14px; line-height: 1.7; margin: 8px 0; padding: 6px;">Use <strong>statsmodels</strong> OLS to estimate regression models. Start with base model (interest rate only), then add controls (market, consumer confidence, disposable income, inflation). Use HC3 robust standard errors to address potential heteroskedasticity.</p>
<h4 style="margin: 15px 0 10px 0; font-size: 15px; font-weight: 600;">Step 5: Visualization & Diagnostics</h4>
<p style="font-size: 14px; line-height: 1.7; margin: 8px 0; padding: 6px;">Use <strong>matplotlib</strong> to create visualizations (time series, scatter plots, residuals, Q-Q plots). Use <strong>statsmodels</strong> diagnostic functions to test model assumptions (VIF, Breusch-Pagan, Durbin-Watson, Jarque-Bera).</p>
<p style="font-size: 14px; line-height: 1.7; margin: 20px 0; padding: 6px; text-align: center;" ><strong>Together, these tools form a complete research pipeline</strong> from data collection to results presentation, enabling reproducible and rigorous analysis.</p>
</div>

Rising inflation may signal future monetary tightening, affecting BNPL firms through both funding costs and consumer spending patterns. This finding suggests BNPL firms are sensitive to both current and expected monetary policy.</p>

While not statistically significant (p = 0.202), the economic magnitude is substantial and aligns with: (1) theoretical predictions about BNPL firms' structural vulnerabilities, (2) observed valuation patterns (e.g., Klarna's 71% decline), and (3) firm-level evidence (Affirm's 394% funding cost increase). The lack of statistical significance reflects high volatility and limited sample size, but the economic magnitude is meaningful.</p>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;" >Limitations & Scope Constraints</h2>
<div style="max-width: 65%; margin: 0 auto; padding: 15px 0;">
<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Sample Size Constraints</h3>
<p style="font-size: 14px; line-height: 1.7; margin: 10px 0;"><strong>Limited Observations:</strong> 67 monthly observations with only ~15-20 non-zero Federal Funds Rate changes reduces statistical power. BNPL firms are relatively new (Affirm IPO in 2021, Sezzle in 2020), limiting historical data availability. This constraint is inherent to studying emerging sectors.</p>
<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Time Period Limitations</h3>
<p style="font-size: 14px; line-height: 1.7; margin: 10px 0;"><strong>COVID-19 Period:</strong> Sample includes pandemic period (2020-2021) with extreme volatility and unconventional monetary policy, potentially confounding results. However, this period also provides valuable policy variation. <strong>Single Rate Cycle:</strong> Sample captures primarily one rate cycle (easing 2020-2021, tightening 2022-2024), limiting generalizability across different monetary regimes.</p>
<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Omitted Variables</h3>
<ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li><strong>Regulatory Changes:</strong> CFPB's May 2024 ruling classifying BNPL as credit cards may have affected stock prices independently of rates</li><li><strong>Competitive Dynamics:</strong> Entry of new BNPL providers and merchant fee competition may affect returns independently</li><li><strong>Firm-Specific Events:</strong> Earnings announcements, management changes, and product launches create noise in portfolio returns</li></ul>
<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Portfolio Construction</h3>
<p style="font-size: 14px; line-height: 1.7; margin: 10px 0;"><strong>Equal Weighting:</strong> Equally-weighted portfolio may not reflect market capitalization weights, though this is standard for small portfolios. <strong>Firm Heterogeneity:</strong> Portfolio approach masks firm-level differences (PayPal is diversified, Affirm is pure-play, Sezzle is smaller), though this is addressed through firm-level analysis.</p>
<h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Causality Limitations</h3>
<p style="font-size: 14px; line-height: 1.7; margin: 10px 0;"><strong>Observational Data:</strong> Regression cannot establish causality from observational data alone. Interest rate changes are not random experiments; they respond to economic conditions that also affect BNPL returns. Results should be interpreted as associations rather than causal effects, though the economic mechanism is well-supported by firm-level evidence.</p>
</div>
