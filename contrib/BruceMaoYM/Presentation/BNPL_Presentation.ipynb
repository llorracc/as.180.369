<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">The BNPL Paradox: Explosive Growth, Valuation Collapse</h2><div style="max-width: 95%; margin: 0 auto; padding: 20px 0;"><table style="width: 90%; margin: 15px auto; border-collapse: collapse; font-size: 13px;"><tr><td style="padding: 6px; border: 1px solid #dee2e6;"><img src="presentation_images/bnpl_gmv_growth_chart.png" alt="BNPL GMV Growth Chart" style="max-width: 100%; height: auto;" /></td><td style="padding: 6px; border: 1px solid #dee2e6;"><img src="presentation_images/klarna_valuation_chart.png" alt="Klarna Valuation Chart" style="max-width: 100%; height: auto;" /></td></tr></table><div style="margin-top: 25px;"><p style="font-size: 14px; line-height: 1.7; margin: 15px 0;">Between 2019 and 2022, the Buy-Now-Pay-Later market exploded: GMV surged from <strong>&#36;2.0B</strong> to <strong>&#36;34.0B</strong>, a <strong>17x increase</strong> in just 3 years. Consumer adoption jumped from 5% to 21% of US consumers, with 61% of borrowers having subprime credit scores. This growth was fueled by zero-interest financing during near-zero Fed rates.</p><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">Yet here's the paradox: despite continued operational growth, BNPL valuations collapsed. Klarna, once valued at <strong>&#36;46B</strong> in June 2021, saw its valuation plummet to <strong>&#36;6.7B</strong> by March 2022, an <strong>85% decline</strong> in just 9 months as the Fed began raising rates. By its 2025 IPO, Klarna was worth <strong>&#36;13.5B</strong>, still <strong>71% below</strong> its peak despite market recovery. This occurred while GMV continued growing, revealing investors priced in interest rate vulnerability that operational metrics couldn't capture.</p><p style="font-size: 14px; line-height: 1.7; margin: 15px 0;">Why did equity markets collapse BNPL valuations during monetary tightening, even as transaction volumes grew? This research examines whether stock returns reflect structural vulnerabilities that could challenge monetary policy transmission as BNPL represents a growing share of consumer credit.</p></div></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Literature Foundation</h2><div style="max-width: 85%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; margin: 15px 0; text-align: center;">The Federal Reserve's ability to influence consumer spending assumes consumer credit responds to rate changes. BNPL challenges this assumption through its unique business model and structural vulnerabilities.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">The Monetary Policy Transmission Problem</h3><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">Traditional monetary policy works through the credit channel: when the Fed raises rates, banks charge higher interest, reducing consumer spending. BNPL firms operate differently: zero-interest to consumers, wholesale funding (immediate pass-through), and razor-thin profit margins (~1%) that can be eliminated by small funding cost increases (CFPB, 2022).</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Structural Vulnerabilities: Evidence from Literature</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Laudenbach (2025) shows each 1pp rate increase raises BNPL funding costs by 0.8-1.0pp, potentially eliminating profit margins. Unlike banks with deposit buffers, BNPL firms face immediate pass-through. CFPB (2023) finds BNPL borrowers have subprime scores (580-669), high credit utilization (60-66%), and 63% originate multiple simultaneous loans. BNPL firms rely on short-term wholesale funding rather than sticky deposits, with Affirm maintaining high debt-to-asset ratios (70.40%), amplifying funding cost impacts on equity returns.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Consumer Behavior & Monetary Policy Effectiveness</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Di Maggio et al. (2022) find BNPL access increases spending by ~&#36;130/week with 24-week persistence, as consumers treat BNPL as "free money." Bian et al. (2023) show BNPL boosts consumption by 15-20%, particularly for liquidity-constrained households. BNPL may reduce monetary policy effectiveness, as consumers may not reduce spending when rates rise if BNPL remains interest-free.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600; color: #0066cc;">Research Gap</h3><p style="color: #0066cc; font-size: 14px; line-height: 1.7; margin: 10px 0;">While existing literature establishes BNPL firms' structural vulnerability to interest rate changes through funding cost analysis, there is limited empirical work directly linking BNPL firm performance to monetary policy through stock return analysis. This research fills that gap by examining how equity markets price interest rate risk for BNPL firms.</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Research Objective & Approach</h2><div style="max-width: 85%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; margin: 15px 0; text-align: center;">To examine how equity markets price interest rate risk for BNPL firms, this research addresses:</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Research Question</h3><p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 12px 0;"><em>How do BNPL firms' stock returns respond to Federal Funds Rate changes?</em></p><p style="font-size: 14px; line-height: 1.7; margin: 15px 0;">If BNPL firms are highly sensitive to rate changes, monetary policy may be less effective at controlling consumer spending in an economy where BNPL represents a growing share of credit. This question matters because understanding BNPL's sensitivity to monetary policy is crucial for assessing the effectiveness of traditional monetary transmission mechanisms in an evolving credit landscape.</p><h3 style="text-align: center; margin: 30px 0 15px 0; font-size: 16px; font-weight: 600;">Methodology</h3><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">This research empirically tests how BNPL firms' stock returns respond to Federal Reserve monetary policy changes using monthly data from February 2020 to August 2025. The analysis employs multi-factor regression to isolate interest rate effects from other economic factors. The portfolio approach captures sector-wide patterns while controlling for market movements, consumer confidence, disposable income, and inflation. This methodology allows us to test whether equity markets price in the structural vulnerabilities that theory predicts.</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Research Hypothesis & Portfolio Construction</h2><div style="max-width: 90%; margin: 0 auto; padding: 15px 0;"><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">BNPL Portfolio Construction</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">BNPL Returns are calculated as an equally-weighted portfolio of three publicly traded BNPL companies: PayPal (PYPL), Affirm (AFRM), and Sezzle (SEZL). This mechanism should be stronger for BNPL firms than traditional banks due to their structural vulnerabilities.</p><p style="font-size: 14px; line-height: 1.7; margin: 12px 0;">Given BNPL firms' reliance on short-term wholesale funding and razor-thin profit margins, I expect β₁ &lt; 0: negative relationship between interest rate changes and BNPL stock returns. The economic magnitude should be substantial given that funding costs can eliminate profit margins entirely.</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Research Tools & Process</h2><div style="max-width: 85%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 15px 0;">This research uses Python libraries to collect data, analyze relationships, and visualize findings through a systematic 5-step process.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Core Tools</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;"><strong>yfinance & fredapi:</strong> Data collection for stock prices and macroeconomic variables. <strong>pandas:</strong> Data manipulation and time series alignment. <strong>statsmodels:</strong> OLS regression with HC3 robust standard errors. <strong>matplotlib:</strong> Visualization for charts and diagnostics.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">5-Step Research Process</h3><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;"><strong>Step 1 - Data Collection:</strong> Download stock prices and macro variables using yfinance and fredapi. pandas handles missing values and time series alignment.</p><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;"><strong>Step 2 - Portfolio Construction:</strong> Calculate equally-weighted BNPL portfolio returns. Convert daily prices to monthly returns using pandas resampling.</p><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;"><strong>Step 3 - Variable Construction:</strong> Transform variables using pandas: log returns, first differences, percentage changes. This ensures stationarity and interpretability.</p><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;"><strong>Step 4 - Model Estimation:</strong> Use statsmodels OLS to estimate regression models. Start with base model, then add controls. Use HC3 robust standard errors.</p><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;"><strong>Step 5 - Visualization & Diagnostics:</strong> Use matplotlib for visualizations. Use statsmodels diagnostic functions to test model assumptions (VIF, Breusch-Pagan, Durbin-Watson, Jarque-Bera).</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Model Specification</h2><div style="max-width: 65%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 12px 0;">Multi-factor regression framework to estimate how BNPL stock returns respond to Federal Funds Rate changes, controlling for other economic factors.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Base Model</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Tests the direct relationship between BNPL returns and interest rate changes, without controls. This provides a baseline estimate of interest rate sensitivity.</p><div style="padding: 6px; margin: 12px 0; text-align: center;"><p style="font-size: 14px; line-height: 1.7; margin: 5px 0; font-family: 'Courier New', monospace;">log(1 + BNPL_Return<sub>t</sub>/100) = β₀ + β₁(∆Federal_Funds_Rate<sub>t</sub>) + ε<sub>t</sub></p></div><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;">Log transformation handles negative returns and makes coefficients interpretable as percentage changes. First differences ensure stationarity for interest rates.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Full Model</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Adds control variables to isolate the "pure" interest rate effect from other economic factors.</p><div style="padding: 6px; margin: 12px 0; text-align: center;"><p style="font-size: 14px; line-height: 1.7; margin: 5px 0; font-family: 'Courier New', monospace;">log(1 + BNPL_Return<sub>t</sub>/100) = β₀ + β₁(∆FFR<sub>t</sub>) + β₂(∆CC<sub>t</sub>) +<br/>β₃(∆DI<sub>t</sub>) + β₄(∆π<sub>t</sub>) + β₅(R<sub>MKT,t</sub>) + ε<sub>t</sub></p></div><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;">Control variables: Market return (systematic risk), Consumer Confidence (demand effects), Disposable Income (spending capacity), Inflation (real vs nominal effects).</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Model Implementation</h2><div style="max-width: 65%; margin: 0 auto; padding: 15px 0;"><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Variable Construction (pandas)</h3><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li>BNPL Returns: <code>np.log1p(returns / 100)</code> to handle negative returns and make coefficients interpretable</li><li>∆FFR: <code>df['dFFR'] = df['FFR'].diff()</code> for first differences to ensure stationarity</li><li>∆CC, ∆DI, ∆π: <code>.diff()</code> for indices, <code>.pct_change()</code> for levels</li><li>R<sub>MKT</sub>: S&P 500 returns calculated same as BNPL returns</li></ul><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Estimation Method (statsmodels)</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Using <code>sm.OLS(y, X).fit(cov_type='HC3')</code> to estimate coefficients with HC3 robust standard errors. However, this period also provides valuable policy variation. Single Rate Cycle: Sample captures primarily one rate cycle, limiting generalizability across different monetary regimes.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Omitted Variables</h3><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li><strong>Regulatory Changes:</strong> CFPB's May 2024 ruling classifying BNPL as credit cards may have affected stock prices independently</li><li><strong>Competitive Dynamics:</strong> Entry of new BNPL providers and merchant fee competition may affect returns</li><li><strong>Firm-Specific Events:</strong> Earnings announcements, management changes, and product launches create noise</li></ul><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Portfolio Construction</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Equal Weighting: Equally-weighted portfolio may not reflect market capitalization weights, though standard for small portfolios. Firm Heterogeneity: Portfolio approach masks firm-level differences (PayPal diversified, Affirm pure-play, Sezzle smaller), though addressed through firm-level analysis.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;" >Causality Limitations</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Observational Data: Regression cannot establish causality from observational data alone. Interest rate changes respond to economic conditions that also affect BNPL returns. Results should be interpreted as associations rather than causal effects, though the economic mechanism is well-supported by firm-level evidence.</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Conclusions</h2><div style="max-width: 65%; margin: 0 auto; padding: 15px 0;"><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Main Finding</h3><p style="font-size: 14px; line-height: 1.7; margin: 15px 0; padding: 6px; text-align: center;">BNPL stock returns show substantial economic sensitivity to interest rate changes (<strong>β₁ = -12.68</strong>).
