<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Model Specification</h2><div style="max-width: 65%; margin: 0 auto; padding: 15px 0;"><p style="font-size: 14px; line-height: 1.7; text-align: center; margin: 12px 0;">Multi-factor regression framework to estimate how BNPL stock returns respond to Federal Funds Rate changes, controlling for other economic factors.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Base Model</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Tests the direct relationship between BNPL returns and interest rate changes, without controls. This provides a baseline estimate of interest rate sensitivity.</p><div style="padding: 6px; margin: 12px 0; text-align: center;"><p style="font-size: 14px; line-height: 1.7; margin: 5px 0; font-family: 'Courier New', monospace;">log(1 + BNPL_Return<sub>t</sub>/100) = β₀ + β₁(∆Federal_Funds_Rate<sub>t</sub>) + ε<sub>t</sub></p></div><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;">Log transformation handles negative returns and makes coefficients interpretable as percentage changes. First differences ensure stationarity for interest rates.</p><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Full Model</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0;">Adds control variables to isolate the "pure" interest rate effect from other economic factors.</p><div style="padding: 6px; margin: 12px 0; text-align: center;"><p style="font-size: 14px; line-height: 1.7; margin: 5px 0; font-family: 'Courier New', monospace;">log(1 + BNPL_Return<sub>t</sub>/100) = β₀ + β₁(∆FFR<sub>t</sub>) + β₂(∆CC<sub>t</sub>) +<br/>β₃(∆DI<sub>t</sub>) + β₄(∆π<sub>t</sub>) + β₅(R<sub>MKT,t</sub>) + ε<sub>t</sub></p></div><p style="font-size: 14px; line-height: 1.7; margin: 8px 0;">Control variables: Market return (systematic risk), Consumer Confidence (demand effects), Disposable Income (spending capacity), Inflation (real vs nominal effects).</p></div>

<h2 style="text-align: center; font-size: 24px; font-weight: 700; margin: 30px 0 20px 0;">Model Implementation</h2><div style="max-width: 65%; margin: 0 auto; padding: 15px 0;"><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Variable Construction (pandas)</h3><ul style="font-size: 14px; line-height: 1.7; margin: 10px 0; padding-left: 22px;"><li>BNPL Returns: <code>np.log1p(returns / 100)</code> to handle negative returns and make coefficients interpretable</li><li>∆FFR: <code>df['dFFR'] = df['FFR'].diff()</code> for first differences to ensure stationarity</li><li>∆CC, ∆DI, ∆π: <code>.diff()</code> for indices, <code>.pct_change()</code> for levels</li><li>R<sub>MKT</sub>: S&P 500 returns calculated same as BNPL returns</li></ul><h3 style="text-align: center; margin: 20px 0 12px 0; font-size: 16px; font-weight: 600;">Estimation Method (statsmodels)</h3><p style="font-size: 14px; line-height: 1.7; margin: 10px 0; margin: 12px 0;">Using <code>sm.OLS(y, X).fit(cov_type='HC3')</code> to estimate coefficients with HC3 robust standard errors.
