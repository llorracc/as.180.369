<h4 style="text-align: center;">The University of Michigan Index of Consumer Sentiment (ICS)</h4>

- Consumer sentiment about finances help economists predict Consumer Spending

- In 1946, George Katona: subjective perceptions + expectations play role in consumer behavior
- created the University of Michigan Survey of Consumers that releases monthly ICS!


  

- UM phone interviews > 600 Americans each month
- Interviewed groups designed to statistically represent all American households
- Primarily, the survey focuses on how people view:
    1. their own financial situation: near + long term
    2. prospects for general economy over near term
    3. prospects for general economy over long term


<h3 style="text-align: center;">But Can Sentiment Actually Predict Sending? </h3>

- Carroll(1994): ICS explains 14% of variation in consumption growth

- Ludvigson(2004): one-point rise in ICS = 1.3% increase in personal consumption

- Curtin(2019): high correlation with interest rate, inflation, and unemployment shifts



<h3 style="text-align: center;">Economic Models Tell Us Econ. Agents Are Rational...But Are Consumers?</h3>

- Most research indicates they aren't

- little/no econ. consequences from Iraqi invasion of Kuwait, no econ. consequences 
- sentiment fell steeply => "animal spirits" of consumer panic (Blanchard(1993))
- This drop was a large contributor to the 1990-1991 recession

- Hall(1993) showed the steep drop in sentiment + consumption was unwarranted considering actual changes to disposable incomes


<h3 style="text-align: center;">Also, Negative News Coverage Causes Lower Sentiment Even, When Times Are Good</h3>

- Doms(2004) developed an R-Word Index of words like "recession", "economic slowdown", "economic crisis", etc.
- searched through major U.S. newspapers and magazines for these words
- found that consumers receive sentiment signalling from the tone  volume of negative economic reporting
- in the 1990's especially, econ. news was much more negative than economic indicators warranted => irrational recession?
- Hollanders(2011) found the same dynamic in Europe and Shapiro(2022) showed this using machine learning


<h3 style="text-align: center;">And People Let Their Political Biases Shape Their Economic Outlook</h3>

- De Boef(2004) found Democrats reported lower consumer sentiment after election of President Bush in 2000
- this, despite no notable changes in macroeconomic indicators
- Post 2016 election, Republicans (Democrats) increased (decresed) equity share in financial portfolios
- => investors signals through lens of political identity, even if asset prices don't reflect their views
- Gillitzer(2018) looked at car sales + large household in Australia post-elections
- Australian consumers report > economic sentiment when their preferred political party is in power in a way that unexplainable by econ. fundamentals

<h3 style="text-align: center;">Since 9-11, Terrorism Has Made Consumers Irrational About the Economy</h3>

- duringSecond Intifada, even perceived potential for future terrorist attacks lowered consumer sentiment =>depressed consumer spending (Waxman(2011))
- this, even though the terrorism rarely had significant economic consequences
- and, Drakos(2015) showed terrorism activity has a statistically significant negative coefficient in a model estimating consumer confidence
- Interstingly, this finding only applies post-9/11 => has terrorism taken on a new, larger role in our economic perceptions?

<h3 style="text-align: center;">We Found No Correlation Between Republican Presidents and Consumer Sentiment </h3>

- for our regressions of GOP Presidents with consumer sentiment
- 1. very low or negative adjusted R-squared
  2. low or negative coefficients
  3. high P values
  4. confidence intervals that cross 0
This makes sense though:
  1. the ICS tries to get a representative sample of American households=> reasonable that the political affiliations of the entire sample doesn't skew in either political direction
  2. Some political researchers have pointed out that there aren't fundamentally deep practical differences in the economic policies pursued by many presidents
  3. For example: economy did very well under Clinton-a social democrat even though he didnt pursue significantly different policies than other dems
