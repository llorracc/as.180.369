# California's Film Tax Credit Programs
## Do They Generate Employment and Wage Gains?

Evaluating Program 2.0 (2015) and Program 3.0 (2020)


## Research Question & Context

**Did California's Film Tax Credit Program 2.0 (2015) and Program 3.0 (2020) generate sustained employment or wage gains?**

- **2009:** First program ($100M/year)  
- **2015:** Program 2.0 - **$330M/year** (largest single-state program)  
- **2020:** Program 3.0 - **$1.55B over 5 years**

**Context:** Film tax credits persist despite weak evidence; 30+ states, $2B+ annually


## Methodology

**Three Complementary Approaches:**

1. **Synthetic Control Method (SCM)** - Preferred method
   - Donor pool: NY, GA, LA, FL, IL, PA
   - No parallel trends assumption required

2. **Difference-in-Differences (DiD)**
   - California vs. Georgia & New York
   - Provides wage estimates

3. **Migration Validation (ACS)**
   - IPUMS ACS microdata (2009-2023)
   - Tests if employment gains reflect real worker relocation

**Data:** BLS QCEW (2010-2025), NAICS 512110


## SCM Results: Employment Effects

| Treatment | Effect | P-value | Rank |
|-----------|--------|---------|------|
| Program 2.0 (2015) | **+3.5%** | 0.29 | 2nd/7 |
| Program 3.0 (2020) | **+4.3%** | 0.14 | 1st/7 |

**Key Finding:** Positive point estimates but **neither statistically significant** (p > 0.05)

<img src="../Data%20Exploration/SCM/scm_results_2020_employment.png" style="max-width: 85%; max-height: 50vh; object-fit: contain;">

**Cannot confidently conclude employment growth**


## Wage Effects (DiD)

| Treatment | Effect | P-value | Significant? |
|-----------|--------|---------|-------------|
| Program 2.0 (2015) | **+9.1%** | **0.003** | **Yes*** |
| Program 3.0 (2020) | -6.9% | 0.15 | No |

**Key Finding:** **Substantial and statistically significant wage premium** following 2015 expansion

**Interpretation:** Policy improved compensation for existing workers, but 2020 effect likely confounded by COVID-19


## Migration Validation (ACS)

**Question:** Do employment gains reflect real worker relocation?

<img src="../Data%20Exploration/Migration/california_film_migration.png" style="max-width: 85%; max-height: 50vh; object-fit: contain;">

**Key Findings:**
- **Persistent net inflow:** +2,029 workers/year (2009-2023)
- **2015:** +2,058 workers (+1.081%)
- **2020:** +1,667 workers (+0.737%)
- **2021:** +58 workers (+0.028%) - **smallest inflow**

**Net inflow did not strengthen after tax credits**


## Integrated Findings

### Employment Effects (SCM)
- Program 2.0: **+3.5%** (p = 0.29) ⚠️ **Not significant**
- Program 3.0: **+4.3%** (p = 0.14) ⚠️ **Not significant**

### Wage Effects (DiD)
- Program 2.0: **+9.1%** (p = 0.003) ⭐ **Significant**

### Migration (ACS)
- Persistent net inflow (+2,029/year) but **did not strengthen** after credits
- 2021: smallest inflow (+58 workers)

**Conclusion:** No statistically significant employment growth. Substantial wage improvements for existing workers. Net inflow pattern unchanged.


## Interpretation: Three Mechanisms

1. **Baseline Inflow Pattern**
   - Net inflow reflects California's existing attractiveness, not policy effects
   - Tax credits did not enhance baseline pattern

2. **More Work for Existing Workers**
   - Employment gains may reflect longer hours, more consistent work
   - **Wage premium (+9.1%)** supports this

3. **Temporary vs. Permanent Migration**
   - ACS captures permanent moves but misses temporary relocations
   - Project-based work may not show in migration data

**Policy improved compensation but not detectable employment growth**


## Political Timing

| Program | Enactment | Election | Months Before |
|---------|-----------|----------|---------------|
| Program 2.0 | Sep 18, 2014 | Nov 4, 2014 | ~1.5 months |
| Program 3.0 | Jun 27, 2018 | Nov 6, 2018 | ~4.5 months |

**Both enacted in election years, within 6 months of Election Day**

**Probability by chance: ~1.6%**

**Interpretation:** Expansions timed around elections, suggesting political rather than purely economic drivers


## Conclusion & Policy Implications

**Main Findings:**
- **No statistically significant employment growth** (3-4% point estimates, p = 0.29, 0.14)
- **Substantial wage improvements** (+9.1% for 2015, p = 0.003)
- **Net inflow pattern did not strengthen** after tax credits

**Policy Implications:**
- Benefits accrue to existing workers (wage premiums)
- No significant net job creation
- Tax credits insufficient to fundamentally alter competitive position
- Political timing suggests electoral incentives drive expansion

**The policy achieved wage improvements but not detectable employment effects.**
