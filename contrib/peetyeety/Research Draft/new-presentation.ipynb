# California's Film Tax Credit Programs
## Do They Generate Employment and Wage Gains?

Evaluating Program 2.0 (2015) and Program 3.0 (2020)


## Research Question

**Did California's Film Tax Credit Programs generate sustained employment or wage gains?**

| Program | Year | Annual Allocation |
|---------|------|-------------------|
| Program 1.0 | 2009 | 100 million |
| Program 2.0 | 2015 | **330 million** (largest in nation) |
| Program 3.0 | 2020 | **1.55 billion** over 5 years |

**Policy Puzzle:** 30+ states spend 2+ billion annually on film incentives despite weak evidence of economic benefits


## Literature Review

<div style="display: flex; align-items: center;">
<div style="flex: 1; padding-right: 20px;">

**Three Strands:**

1. **Creative-Economy Narratives** (Christopherson & Rightor 2010, Adkisson 2013)
2. **National Evidence** (Thom 2018, Button 2019, Bradbury 2020)
3. **State-Level & Political** (Thom 2018b, Rickman & Wang 2020-2022, Owens & Rennhoff 2024)

**Consistent finding:** Modest effects, limited evidence of broader development

## Methodology

**Three Complementary Approaches:**

1. **Synthetic Control Method (SCM)** - Preferred method
   - Donor pool: NY, GA, LA, FL, IL, PA
   - No parallel trends assumption required

2. **Difference-in-Differences (DiD)**
   - California vs. Georgia & New York
   - Provides wage estimates

3. **Migration Validation (ACS)**
   - IPUMS ACS microdata (2009-2023)
   - Tests if employment gains reflect real worker relocation

**Data:** BLS QCEW (2010-2025), NAICS 512110


## DiD Results: Employment & Wages

### Employment Effects
| Treatment | Effect | P-value | Significant? |
|-----------|--------|---------|-------------|
| Program 2.0 (2015) | **-18.3%** | 0.316 | No |
| Program 3.0 (2020) | **-3.5%** | 0.526 | No |

### Wage Effects
| Treatment | Effect | P-value | Significant? |
|-----------|--------|---------|-------------|
| Program 2.0 (2015) | **+9.1%** | **0.003** | **Yes*** |
| Program 3.0 (2020) | -6.9% | 0.15 | No |

**Key Finding:** ⚠️ DiD suffers from parallel trends violation. **Wage premium (+9.1%) is significant** but employment estimates are biased.

## Migration Validation (ACS)

<div style="display: flex; align-items: flex-start;">
<div style="flex: 1; padding-right: 20px;">

**Question:** Do employment gains reflect real worker relocation?

**Key Findings:**
- **Persistent net inflow:** +2,029 workers/year (2009-2023)
- **2015:** +2,058 workers (+1.081%)
- **2020:** +1,667 workers (+0.737%)
- **2021:** +58 workers (+0.028%) - **smallest inflow**

**Net inflow did not strengthen after tax credits**

## Political Timing

| Program | Signed | Election | Before |
|---------|--------|----------|--------|
| 2.0 | Sep 18, 2014 | Nov 4, 2014 | 1.5 mo |
| 3.0 | Jun 27, 2018 | Nov 6, 2018 | 4.5 mo |

**Both enacted within 6 months of gubernatorial elections**

Probability if random: **(0.125)² ≈ 1.6%**

Suggests electoral rather than economic motivations


## Conclusion & Policy Implications

**Main Findings:**
- **No statistically significant employment growth** (SCM: 3-4% point estimates, p = 0.29, 0.14)
- **Substantial wage improvements** (DiD: +9.1% for 2015, p = 0.003)
- **Net inflow pattern did not strengthen** after tax credits

**Policy Implications:**
- Benefits accrue to existing workers (wage premiums)
- No significant net job creation
- Tax credits insufficient to fundamentally alter competitive position
- Political timing suggests electoral incentives drive expansion

**The policy achieved wage improvements but not detectable employment effects.**
