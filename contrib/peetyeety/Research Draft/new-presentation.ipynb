# California's Film Tax Credit Programs
## Do They Generate Employment and Wage Gains?

**Peter Li** | Senior Thesis | Economics


## Research Question

**Did California's Film Tax Credit Programs generate sustained employment or wage gains?**

| Program | Year | Annual Allocation |
|---------|------|-------------------|
| Program 1.0 | 2009 | 100 million |
| Program 2.0 | 2015 | **330 million** (largest in nation) |
| Program 3.0 | 2020 | **1.55 billion** over 5 years |

**Policy Puzzle:** 30+ states spend 2+ billion annually on film incentives despite weak evidence of economic benefits


## Literature Review

<div style="display: flex; align-items: flex-start;">
<div style="flex: 0.9; padding-right: 15px;">

**Prior Research Finds:**

- Film incentives rarely generate lasting employment gains (Thom 2018; Button 2019; Bradbury 2020)
- Programs persist for political reasons (Owens & Rennhoff 2023)

**Gap:** No analysis of California's 2015/2020 expansions

## Methodology

| Method | Purpose | Data |
|--------|---------|------|
| **Synthetic Control (SCM)** | Employment effects | QCEW 2010-2025 |
| **Difference-in-Differences** | Wage effects | QCEW 2010-2025 |
| **Migration Validation** | Real worker flows? | ACS 2009-2023 |
| **Political Timing** | Electoral incentives | Legislative records |

**Control States:** New York, Georgia (stable 30% credits)

**Industry:** NAICS 512110 (Motion Picture Production)


## DiD Results

| Outcome | Program 2.0 (2015) | Program 3.0 (2020) |
|---------|-------------------|-------------------|
| **Employment** | -18.3% (p=0.316) | -3.5% (p=0.526) |
| **Wages** | **+9.1% (p=0.003)** ✓ | -6.9% (p=0.153) |

**⚠️ Limitation:** DiD fails parallel trends test — employment estimates unreliable

**Key Finding:** Significant **+9.1% wage premium** (~11,000 dollars/worker annually)

## Political Timing

| Program | Signed | Election | Before |
|---------|--------|----------|--------|
| 2.0 | Sep 18, 2014 | Nov 4, 2014 | 1.5 mo |
| 3.0 | Jun 27, 2018 | Nov 6, 2018 | 4.5 mo |

**Both enacted within 6 months of gubernatorial elections**

Probability if random: **(0.125)² ≈ 1.6%**

Suggests electoral rather than economic motivations


## Conclusion

| Finding | Result |
|---------|--------|
| **Employment** | No significant effect (p = 0.286, 0.143) |
| **Wages** | **+9.1% premium** (p = 0.003) |
| **Migration** | No post-policy increase |
| **Political Timing** | Both in election years (~1.6% if random) |

**Interpretation:** Tax credits function as **defensive policy** — improving wages for existing workers rather than generating job creation

**Broader Implication:** If this well-funded program can't detectably expand employment, states should question whether incentives outperform infrastructure or education investments
