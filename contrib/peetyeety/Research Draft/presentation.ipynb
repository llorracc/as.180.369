**Limitation:** 
- DiD fails parallel trends test (CA vs GA: 0.153 log points/year)
- Placebo test (2013): spurious effect (p=0.005) — employment estimates unreliable

**Key Finding:** 
- Significant **+9.1% wage premium** after Program 2.0
- ≈ 11,000 dollars/worker annually (~1.2 billion aggregate)

**Key Insights:**

**Defensive Policy:** Tax credits improve wages for existing workers rather than creating jobs

**Statistical Power:** 3-4% effects (~3,800 jobs) may exist but lack power to detect

**"Race to the Bottom":** If California's well-funded program can't expand employment, states should reconsider incentives vs. infrastructure/education investments

**Policy Persistence:** Programs continue for political visibility despite null employment effects
