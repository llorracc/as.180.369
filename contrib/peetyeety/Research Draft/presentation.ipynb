# California's Film Tax Credit Programs
## Do They Generate Employment and Wage Gains?

**Research Presentation**

---

Evaluating Program 2.0 (2015) and Program 3.0 (2020)

Using Difference-in-Differences, Synthetic Control Methods, and Migration Validation


## The Policy Puzzle

- **Film tax credits persist** despite decades of research showing limited effects
- Over 30 states offer incentives, **$2+ billion annually** nationwide
- Literature consistently finds: **minimal sustained employment gains**, no aggregate growth

**Why do these programs continue to expand?**


## California's Entry: Late but Large-Scale

### Timeline

- **2009:** First program ($100M annually)
- **2015:** Program 2.0 - **Tripled to $330M** (largest single-state program)
- **2020:** Program 3.0 - Expanded to **$1.55B over 5 years**

### Context

- Streaming boom created unprecedented content demand
- Production technologies enabled location flexibility
- California's share of national employment declining

**Critical test case:** Can large-scale, modernized programs achieve their goals?


## Research Question

**Did California's Film Tax Credit Program 2.0 (2015) and Program 3.0 (2020) generate sustained employment or wage gains?**

### Why This Matters

- Extends Thom (2018) beyond 2013 to capture California's major expansions
- Addresses Rickman & Wang's (2020) "reclassification problem"
- Tests whether larger-scale programs produce stronger effects
- Evaluates policy in the streaming era


## Literature Review: Key Studies

### Thom (2018): "Lights, Camera, but No Action?"
- National analysis (1998-2013)
- **Finding:** Transferable credits modestly increase employment, but effects are limited
- **Gap:** Ends in 2013, before California's major 2015 expansion

### Rickman & Wang (2020): "Lights, Camera, What Action?"
- Methodological guidance: Use SCM, validate with migration data
- **Concern:** QCEW employment gains may reflect reclassification, not real relocation

### Bradbury (2020)
- **Finding:** No effect on aggregate state economic outcomes

### Owens & Rennhoff (2024)
- **Finding:** Political incentives drive adoption, not economic effectiveness


## Methodology: Three Complementary Approaches

### 1. Difference-in-Differences (DiD)
- Compare California vs. Georgia & New York
- Two treatments: 2015 Q2 and 2020 Q3
- Outcomes: Employment and Wages

### 2. Synthetic Control Method (SCM)
- Construct optimal counterfactual from donor pool
- Donor states: NY, GA, LA, FL, IL, PA
- Outcome: Employment only

### 3. Migration Validation (ACS)
- IPUMS ACS microdata (2009-2023)
- Test if employment gains reflect real worker relocation
- Addresses "reclassification problem"


## Data Sources

### Employment & Wages
- **BLS QCEW** (2010-2025)

- Quarterly data: employment, average weekly wages

- California vs. Georgia: **0.153 log points/year difference** (substantial violation)
- Placebo test fails (p = 0.005)

## DiD Results: Wage Effects

| Treatment | Coefficient | % Change | P-value | Significant? |
|-----------|-------------|----------|---------|-------------|
| Program 2.0 (2015) | +0.087 | **+9.1%** | **0.003** | **Yes*** |
| Program 3.0 (2020) | -0.072 | -6.9% | 0.153 | No |

### Key Finding

**Substantial and statistically significant wage premium following 2015 expansion**

- Suggests meaningful improvement in worker compensation
- May reflect increased demand for skilled workers or improved bargaining power
- 2020 effect likely confounded by COVID-19



## Why SCM? Addressing DiD Limitations

### DiD Problems:
- ❌ Parallel trends violated (Georgia's rapid growth)
- ❌ Placebo test fails
- ❌ Estimates confounded by pre-existing trends

### SCM Advantages:
- ✅ No parallel trends assumption required
- ✅ Optimally weights donor states based on pre-treatment fit
- ✅ Naturally down-weights states with divergent trends
- ✅ More credible counterfactual construction


## SCM Results: Program 2.0 (2015)

![SCM 2015](../Data%20Exploration/SCM/scm_results_2015_employment.png)

**Key Results:**
- **Treatment Effect: +3.5%** employment increase
- **P-value: 0.29** (ranks 2nd out of 7 in placebo tests)
- **Optimal Weight:** New York 100% (best counterfactual match)
- Pre-treatment fit: Excellent (RMSPE = 0.900)

- **Optimal Weight:** New York 99.99%
- Pre-treatment fit: Excellent (RMSPE = 0.918)
- **Approaches marginal statistical significance**

## Migration Validation: The Critical Test

**Question:** Do employment gains reflect real worker relocation?

**Data:** IPUMS ACS microdata (2009-2023)
- Measures permanent state-to-state moves
- Net migration = Inflows - Outflows

**Addresses Rickman & Wang's "reclassification problem"**


## Migration Results: Persistent Net Outflow

![Migration Patterns](../Data%20Exploration/Migration/california_film_migration.png)

**Key Finding:**
- **Net outflow in every year** (2009-2023)
- Average: **-2,029 workers annually** (-1.034% of workforce)
- Total outflow: **-30,428 workers** over study period


## Migration: Treatment Period Patterns

| Year | Net Migration | Migration Rate |
|------|---------------|-----------------|
| **2015** (Program 2.0) | **-2,058** | **-1.081%** |
| 2016 | -1,298 | -0.659% |
| **2020** (Program 3.0) | **-1,667** | **-0.737%** |
| **2021** | **-58** | **-0.028%** | ⭐ **Smallest outflow** |
| 2022 | -1,206 | -0.565% |
| 2023 | -1,878 | -0.843% |

**2021 shows substantial improvement** (temporary)


## The Apparent Contradiction

### SCM Shows:
- ✅ **+3-4% employment increases**
- ✅ Positive effects approaching significance

### ACS Shows:
- ❌ **Persistent net outflow** of workers
- ❌ California losing 2,000+ workers annually

**How can both be true?**


## Reconciling the Evidence: Three Mechanisms

### 1. **Retention of Existing Workers**
- Employment gains reflect **keeping existing workers** rather than attracting new ones
- Tax credits slowed departures without generating net inflows
- Consistent with 2021 data showing smallest outflow (-58 workers)

### 2. **More Work for Existing Workers**
- QCEW measures **jobs**, which can increase through:
  - Longer hours for existing workers
  - More consistent work opportunities
- **Wage premium (+9.1%)** supports this interpretation

### 3. **Temporary vs. Permanent Migration**
- Film production is **project-based**
- ACS captures permanent moves but **misses temporary relocations**
- Employment gains may reflect temporary work without permanent migration


## Integrated Findings Summary

### Employment Effects (SCM - Preferred)
- Program 2.0: **+3.5%** (p = 0.29)
- Program 3.0: **+4.3%** (p = 0.14)

### Wage Effects (DiD)
- Program 2.0: **+9.1%** (p = 0.003) ⭐ **Significant**
- Program 3.0: -6.9% (p = 0.15)

### Migration Patterns (ACS)
- Persistent net outflow: **-2,029 workers/year**
- 2021 improvement: **-58 workers** (smallest outflow)

**Conclusion:** Modest employment gains + substantial wage improvements, but primarily through **retention and better conditions for existing workers**, not net in-migration


## Policy Implications

### What the Policy Achieved:
- ✅ Modest employment increases (3-4%)
- ✅ Substantial wage improvements (+9.1%)
- ✅ Improved retention (2021: smallest outflow)
- ✅ Better conditions for existing workers

### What the Policy Did NOT Achieve:
- ❌ Net in-migration of workers
- ❌ Reversal of long-term outflow trend
- ❌ Fundamental change in competitive position

**Interpretation:** More effective as a **defensive policy** (preventing losses) than as an **offensive strategy** (attracting new production)

## Limitations

### Statistical Power
- Small SCM donor pool (n=6) limits ability to achieve conventional significance
- Effects approach but don't reach p < 0.10 threshold

### Data Constraints
- No program expenditure data → Cannot calculate cost-per-job
- ACS misses temporary migration → May understate production activity effects

### Scope
- Aggregate industry-level analysis
- No heterogeneous effects by occupation, production type, or region

### Methodological Trade-offs
- DiD: Parallel trends violated, but provides wage estimates
- SCM: More credible but limited statistical power

## Conclusion

### Main Findings

California's film tax credit programs produced:
- **Modest employment increases** (3-4%, approaching significance)
- **Substantial wage improvements** (+9.1% for 2015)
- **Improved retention** of existing workers
- **But NOT net in-migration** (persistent outflows continue)

### Interpretation

The policy achieved **partial success** as a defensive strategy—retaining production activity and improving conditions for existing workers—but did **not fundamentally reverse** California's competitive decline.

**Policy functioned more as retention tool than attraction mechanism.**


## Implications for Policymakers

### Key Takeaways

1. **Modest effects persist even at large scale**
   - $1.55B program produces similar effects to smaller programs

2. **Benefits accrue to existing workers**
   - Wage premiums and retention improvements
   - But limited net job creation

3. **Tax credits may be insufficient alone**
   - Other factors (infrastructure, talent, costs) continue to drive location decisions
   - Cannot fundamentally alter competitive position

4. **Cost-effectiveness questions remain**
   - Need expenditure data to evaluate whether benefits justify costs

**The challenge:** Balance visible benefits to existing workers against substantial public costs


## Future Research Directions

1. **Expand donor pool** to improve SCM statistical power

   - By occupation (creative vs. support staff)
