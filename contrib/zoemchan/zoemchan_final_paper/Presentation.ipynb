<h3 style="text-align: center;">The Home Owners Loan Corporation (HOLC) & Redlining</h3>

- HOLC created as part of New Deal effort to stabilize housing markets

- Created “residential security maps” 
    - A (“Best”)
    - B (“Still Desirable”)
    - C (“Definitely Declining”)
    - D (“Hazardous”)


- HOLC’s purpose was to assess mortgage risk, but appraisal practices institutionalized racial prejudice 

<h3 style="text-align: center;">Modern Outcomes</h3>

- Redlined areas exhibit lower home values, lower household incomes, and higher racial segregation

- Also experience higher exposure to air pollution, urban heat, and chronic disease

- Schools in redlined neighborhoods:

   - Receive less per-pupil revenue
   - Enroll more racially homogeneous student bodies
   - Lower average test scores

<h3 style="text-align: center;">To what extent does historical redlining continue to shape access to advanced STEM coursework in U.S. public high schools?</h3>

- Lower property values → fewer school resources → limited capacity to offer advanced STEM

- Segregation → fewer rigorous courses and STEM-qualified teachers in C/D neighborhoods

- Redlining shaped enduring neighborhood opportunity → students inherit unequal access to STEM pathways

<h2 style="text-align: center;">National Level Patterns</h2>

![Figure 1](Figures/Nat_Bar_Chart.png)

- Schools in A-rated areas are far more likely to offer all three courses (44.2%), while only 21.2% of schools in D-rated areas do so
- Conversely, the share of schools offering no advanced STEM subjects increases from 20.9% in A-rated areas to 26.9% in D-rated neighborhoods

![Figure 3](Figures/Nat_Avg_STEM_Classes.png)

- A-rated neighborhoods offer approximately 3.4 calculus classes, 5.6 computer science classes, and more than 8 physics classes.
- D-rated areas offer only about 1 calculus class, 2.2 computer science classes, and 4.3 physics classes.

<h2 style="text-align: center;">Conclusion</h2>

- STEM access declines along HOLC A → D; redlined areas offer fewer advanced STEM courses.

- Patterns persist within boundary buffers → neighborhood disadvantage, not citywide factors, drives gaps.

- City cases vary: Chicago/Detroit show large deficits; Philadelphia smaller; Los Angeles disrupted by magnet/specialty schools.

- Results highlight enduring historical effects → targeted STEM investment can reduce persistent inequities.
