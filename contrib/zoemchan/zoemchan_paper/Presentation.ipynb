<h3 style="text-align: center;">The Home Owners Loan Corporation & Redlining</h3>

- The Home Owners’ Loan Corporation (HOLC) was created as part of the New Deal’s effort to stabilize housing markets

- Created “residential security maps” that graded neighborhoods based on perceived credit risk. 
    - A (“Best”)
    - B (“Still Desirable”)
    - C (“Definitely Declining”)
    - D (“Hazardous”)


- Although the HOLC’s stated purpose was to assess mortgage risk, its appraisal practices institutionalized racial prejudice within federal lending policy

<h3 style="text-align: center;">Modern Outcomes</h3>

- Formerly redlined areas continue to exhibit lower home values, lower household incomes, and higher racial segregation

- These areas also experience higher exposure to air pollution, urban heat, and chronic disease

- Lukes and Cleveland find that schools located in formerly redlined neighborhoods receive significantly less per-pupil revenue, enroll more racially homogeneous student bodies, and have lower average test scores

<h3 style="text-align: center;">To what extent does historical redlining continue to shape access to advanced STEM coursework in U.S. public high schools?</h3>

- Lower property values in redlined areas → fewer school resources → limited capacity to offer advanced STEM

- Segregation from redlining → fewer rigorous courses and STEM-qualified teachers in C/D neighborhoods

- Redlining shaped enduring neighborhood opportunity → students inherit unequal access to STEM pathways
